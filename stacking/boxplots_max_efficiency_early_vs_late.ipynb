# Objective

Compare max efficiency values for Early vs. Middle and Middle/Late jar stacks.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Amphora Stacking 20210925.csv')
```

```python
df
```

```python
df_early = df[df.Period=='Early']
df_middle_late = pd.concat([df[df.Period=='Middle'], df[df.Period=='Middle, Late']])
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=300)
medianprops=dict(color='xkcd:red')
ax.boxplot([df_early['Max Efficiency'], df_middle_late['Max Efficiency']],
           flierprops=flierprops, medianprops=medianprops)
plt.scatter([1]*len(df_early), df_early['Max Efficiency'], facecolor='xkcd:gold')
plt.scatter([2]*len(df_middle_late), df_middle_late['Max Efficiency'], facecolor='xkcd:gold')
plt.xticks(ticks=[1, 2], labels=['Early', 'Middle, Middle/Late'])
plt.title('Max Efficiency')
plt.show()
```
